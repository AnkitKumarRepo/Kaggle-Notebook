# **Lable Encoding**
> Converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
cancer_dataset=pd.read_csv('/kaggle/input/breast-cancer-data/breast_cancer_data.csv')
```

```python
cancer_dataset.head()
```

```python
# finding the count of different labels
cancer_dataset['diagnosis'].value_counts()
```

0 --> Benign

1 --> Malignant

```python
# load the Label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_dataset.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer_dataset['target'] = labels
     
```

```python
cancer_dataset.head()
```

```python
cancer_dataset.sample(10)
```

Label Encoding of iris data

```python
iris_dataset=pd.read_csv('/kaggle/input/iris-dataset/iris_data.csv')
```

```python
iris_dataset.sample(5)
```

```python
iris_dataset['Species'].value_counts()
```

```python
# Loading the Label Encoder
label_iris=LabelEncoder()
```

```python
iris_label=label_iris.fit_transform(iris_dataset.Species)
```

```python
iris_dataset['target']=iris_label
```

```python
iris_dataset.sample(10)
```

```python
# value_count
iris_dataset['target'].value_counts()
```

Iris-setosa --> 0

Iris-versicolor --> 1

Iris-virginica --> 2
