# Imbalanced Dataset:

A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
     
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('/kaggle/input/credit-dataset/credit_data.csv')
```

```python
credit_card_data.sample(10)
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
#separating the legit and fradulent transaction
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
```

```python
print(legit)
```

```python
legit.sample(10)
```

```python
print(fraud.shape)
```

```python
fraud.sample(12)
```

```python
print(fraud)
```

Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 492

```python
legit_sample=legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.sample(11)
```
